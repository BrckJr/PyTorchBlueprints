```python
import torch
from torch import nn
import matplotlib.pyplot as plt

torch.__version__
```

```python
# Setup device agnostic code
# Using the GPU if possible
device = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using device: {device}")
```

## Creating sample data ##

```python
# Creating sample data 
w = -0.7
b = 0.3

# Create data
start = 0
end = 1
step = 0.002

# Use unsqueeze to create a column vector
# Add a second dimension with size 1. This result in a tensor of dimensions ([50,1])
# .unsqueeze(dim=0) would result in dimensions ([1,50])
X = torch.arange(start, end, step).unsqueeze(dim=1)
# Add a randomness to the data
X_dist = X + torch.randn(size=(X.shape))*0.05

y = w * X_dist + b

# Split data in training and test sets with 80 / 20 distribution
# If necessary, split further validation set as well
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
```

```python
# Visualize the data we learn on
def plot_predictions(train_data=X_train, train_labels=y_train, 
                     test_data=X_test, test_labels=y_test, predictions=None):
    """
    Plot the training data, test data and compare the predictions.
    """
    plt.figure(figsize=(10,7))

    # Training data in blue
    plt.scatter(train_data, train_labels, c="b", s=4, label="Training data")

    # Training data in red
    plt.scatter(test_data, test_labels, c="r", s=4, label="Test data")

    if predictions is not None:
        # Plot the predictions in green (predictions on test data)
        plt.scatter(test_data, predictions, c="g", s=4, label="Predictions")

    # Show legend
    plt.legend(prop={"size": 14});
```

```python
plot_predictions()
```

## Building Linear Regression Model Class ##
Option 1: Manually generate the weights and biases

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Initialize the weights of the model at random
        # torch.randn returns random values based on the standard normal distribution (0-mean, 1-variance)
        # requires_grad shows that we can calculate the gradient of these parameters.
        self.weights = nn.Parameter(torch.randn(1), requires_grad=True)

        # Initialize the bias of the model at random
        # torch.randn returns random values based on the standard normal distribution (0-mean, 1-variance)
        # requires_grad shows that we can calculate the gradient of these parameters.   
        self.bias = nn.Parameter(torch.randn(1), requires_grad=True)      
    
    # Forward defines the computation in the model
    # Definition of method ensures the types of input_data (see arguments list) and 
    # return data (see -> torch.Tensor) TYPESAFE!!
    def forward(self, input_data : torch.Tensor) -> torch.Tensor: 
        return self.weights * input_data + self.bias
```

## Generating the Model Instance ##

```python
# Set manual seed as the nn.Parameters are randomly initialized
torch.manual_seed(42)

# Create an instance of the model
model = LinearRegressionModel()

# The model is by default on the CPU, therefore we switch to GPU if possible
model.to(device)

# Check the parameters in the model
list(model.parameters())

# Get the state of the model
model.state_dict()

# Check if the model runs on GPU or CPU
next(model.parameters()).device
```

## Training the Model ##

```python
# Mean Least Squares Error Function
loss_function = torch.nn.MSELoss()

# Stochastic Gradient descent with learning rate 0.01
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
```

```python
torch.manual_seed(42)

# Set the number of epochs (how many times the model will pass over the training data)
epochs = 1000

# Put the data on the available device, GPU if possible
X_train = X_train.to(device)
X_test = X_test.to(device)
y_train = y_train.to(device)
y_test = y_test.to(device)

# Create empty loss lists to track values
train_loss_values = []
test_loss_values = []
epoch_count = []

for epoch in range(epochs):
    """
    Training
    
    """
    # Put model in training mode (this is the default state of a model)
    model.train()

    # 1. Forward pass on train data using the forward() method inside 
    y_pred = model(X_train)
    # print(y_pred)

    # 2. Calculate the loss (how different are our models predictions to the ground truth)
    loss = loss_function(y_pred, y_train)

    # 3. Zero grad of the optimizer
    optimizer.zero_grad()

    # 4. Loss backwards
    loss.backward()

    # 5. Progress the optimizer
    optimizer.step()

    """
    Testing
    
    """
    
    # Put the model in evaluation mode
    model.eval()

    with torch.inference_mode():
      # 1. Forward pass on test data
      test_pred = model(X_test)

      # 2. Caculate loss on test data
      # Predictions are in torch.float, so comparisons need to be done with tensors of the same type
      test_loss = loss_function(test_pred, y_test.type(torch.float)) 

      # Print out what's happening every 20th epoch
      if epoch % 20 == 0:
            train_loss_values.append(loss.cpu().detach().numpy())
            test_loss_values.append(test_loss.cpu().detach().numpy())
            epoch_count.append(epoch)
            if epoch % 50 == 0:
                  print(f"Epoch: {epoch} | MSE Train Loss: {loss} | MSE Test Loss: {test_loss} ")


print("\n", "New State of the model:", "\n")
print(model.state_dict())
```

```python
# Plot the loss curves
plt.plot(epoch_count, train_loss_values, label="Train loss")
plt.plot(epoch_count, test_loss_values, label="Test loss")
plt.title("Training and test loss curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend();
```

## Making Predictions with the Model ##

```python
# 1. Set in evaluation mode
model.eval()

# 2. Make prediction in the inference mode context manager
with torch.inference_mode():
    # 3. Check if model and data are both on the same device. 
    device = "mps" if torch.backends.mps.is_available() else "cpu"
    model.to(device)
    X_test = X_test.to(device)
    y_pred = model(X_test)


# 4. Plot predicitions
plot_predictions(predictions=y_pred.cpu())
```
